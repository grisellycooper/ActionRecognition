```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
from sklearn.manifold import TSNE
```

```python
boxing = pd.read_csv('boxing.csv', sep="|", header=None)
```

```python
dboxing = boxing.values
```

```python
for i, row in enumerate(dboxing):
    mat = np.array([float(d) for d in row[3][:-1].split(',')])
    size = row[2].split(',')
    size = (int(size[0]), int(size[1]))
    np.array(mat)
    mat = mat.reshape(*size)
    row[3] = mat
```

```python
dboxing[30][3].shape
```

```python
def plot_video(index):
    x = dboxing[index]
    features = x[3]
    print("Category:", x[0])
    print("Name:", x[1])
    c, v = [], []
    features = TSNE(n_components=2).fit_transform(features)
    for i, j in features:
        c.append(i)
        v.append(j)
    plt.scatter(c, v)
```

```python
plot_video(4)
```

```python
plot_video(2)
```
